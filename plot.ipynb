```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
betaC = json.loads(open('dctt_betas0_C.json').read())
betaP = json.loads(open('dctt_betas0_P.json').read())
```

```python
alphaC = pd.read_json('dctt_alpha_C.json').transpose()
alphaP = pd.read_json('dctt_alpha_P.json').transpose()
betaC = pd.Series(betaC)
betaP = pd.Series(betaP)
```

```python
betaC
```

```python
def plotAlpha(df, title):
    fig = plt.figure(figsize=(15,8))
    for c in df.columns:
        plt.plot(df[c], label=f"k = {c}")
    plt.legend()
    plt.title(title)
    plt.xlabel("r")
    plt.grid()
    plt.ylabel("Alpha(r)")
    plt.show()
```

```python
def plotBeta(df, title):
    fig = plt.figure(figsize=(15,8))
    plt.plot(df, label=f"m", marker='o')
    # plt.legend()
    plt.title(title)
    plt.xlabel("m")
    plt.grid()
    plt.ylim(0, 1)
    plt.ylabel("Beta(m,0)")
    plt.show()
```

```python
def subplotAlphaBeta(alpha, beta, title):
    fig, axs = plt.subplots(1,2, figsize=(18,7))
    for c in alpha.columns:
        axs[0].plot(alpha[c], label=f"k = {c}")
    
    axs[0].legend()
    axs[0].set_title('Alpha(r) vs r com m = 3')
    axs[0].set_xlabel("r")
    axs[0].grid()
    axs[0].set_ylabel("Alpha(r)")
    axs[0].set_ylim(0, 900)

    axs[1].bar(beta.index, height=beta.values, color = 'tab:orange')
    axs[1].plot(beta.index, beta.values, label=f"m", marker='^', linestyle='--')
    axs[1].set_title('Beta(m, 0) vs m com r = 20 e k = 8')
    axs[1].set_xlabel("m")
    axs[1].grid()
    axs[1].set_ylim(0, 1)
    axs[1].set_ylabel("Beta(m,0)")

    plt.suptitle(title, fontsize=16)
    plt.show()
```

```python
plotAlpha(alphaC, "Alpha(r) vs r na ordenação em Cascata")
```

```python
plotAlpha(alphaP, "Alpha(r) vs r na ordenação Polifásica com m=3")
```

```python
plotBeta(betaC, "Beta(m, 0) vs m na ordenação em Cascata com r=20")
```

```python
plotBeta(betaP, "Beta(m, 0) vs m na ordenação Polifásica com r=20")
```

```python
subplotAlphaBeta(alphaP, betaP, "Ordenação Polifásica")
```

```python
subplotAlphaBeta(alphaC, betaC, "Ordenação em Cascata")
```
