```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
betaC = json.loads(open('dctt_betas0_C.json').read())
betaP = json.loads(open('dctt_betas0_P.json').read())
```

```python
alphaC = pd.read_json('dctt_alpha_C.json').transpose()
alphaP = pd.read_json('dctt_alpha_P.json').transpose()
betaC = pd.Series(betaC)
betaP = pd.Series(betaP)
```

```python
betaC
```

```python
def plotAlpha(df, title):
    fig = plt.figure(figsize=(15,8))
    for c in df.columns:
        plt.plot(df[c], label=f"k = {c}")
    plt.legend()
    plt.title(title)
    plt.xlabel("r")
    plt.grid()
    plt.ylabel("Alpha(r)")
    plt.show()
```

```python
def plotBeta(df, title):
    fig = plt.figure(figsize=(15,8))
    plt.plot(df, label=f"m", marker='o')
    # plt.legend()
    plt.title(title)
    plt.xlabel("m")
    plt.grid()
    plt.ylabel("Beta(m,0)")
    plt.show()
```

```python
plotAlpha(alphaC, "Alpha(r) vs r na ordenação em Cascata")
```

```python
plotAlpha(alphaP, "Alpha(r) vs r na ordenação Polifásica")
```

```python
plotBeta(betaC, "Beta(m, 0) vs m na ordenação em Cascata")
```

```python
plotBeta(betaP, "Beta(m, 0) vs m na ordenação Polifásica")
```
